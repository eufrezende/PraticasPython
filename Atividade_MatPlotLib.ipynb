```python
print("Conjunto 1:")
X = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio"]
Y = [2000, 2500, 3000, 2800, 3500]

width = 0.2
index = np.arange(5)

plt.bar(index, Y, width=width)
plt.xlabel("Mês")
plt.ylabel("Venda")
plt.xticks(index, X)
plt.show()
```

```python
print("Conjunto 1:")
X = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio"]
Y = [2000, 2500, 3000, 2800, 3500]


plt.pie(Y, labels=X, autopct="%.2f%%", shadow=True);
plt.show()
```

```python
print("Conjunto 1:")
mes = ["Janeiro", "Fevereiro", "Março", "Abril", "Maio"]
vendas = [2000, 2500, 3000, 2800, 3500]

sns.barplot(x=mes, y=vendas, hue=mes, palette="viridis")
plt.xlabel('Mês')
plt.ylabel('Vendas')
plt.xticks(rotation=45)
plt.show()
```

```python
print("Conjunto 2:")
data = {"Anos": [2010, 2011, 2012, 2013, 2014, 2015, 2016],
        "População A": [112, 115, 118, 121, 124, 127, 131],
        "População B": [105, 108, 111, 114, 117, 128, 132]}

popl = pd.DataFrame(data)

plt.plot(popl["Anos"], popl["População A"])
plt.plot(popl["Anos"], popl["População B"])
plt.xlabel("Anos")
plt.ylabel("População")
plt.title("Crescimento populacional durante os anos")
plt.legend(["População A (milhões)", "População B (milhões)"])
plt.show()
```

```python
print("Conjunto 2:")
data = {"Anos": [2010, 2011, 2012, 2013, 2014, 2015, 2016],
        "População A": [112, 115, 118, 121, 124, 127, 131],
        "População B": [105, 108, 111, 114, 117, 128, 132]}

popl = pd.DataFrame(data)

width = 0.2
index = np.arange(7)

plt.bar(index, popl['População A'], width=width, label="População A")
plt.bar(index + width, popl['População B'], width=width, label="População B")
plt.xlabel("Ano")
plt.ylabel("População")
plt.xticks(index + width / 2, popl['Anos'])
plt.ylim([0,150])
plt.legend(["População A (milhões)", "População B (milhões)"])
plt.show()
```
